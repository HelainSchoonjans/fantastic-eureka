```python
# there seems to be '-' in some fields instead of NAs...
# let's remove the carets and check if the datatype of the column changed...
import pandas as pd
import numpy as np

df = pd.read_csv("input/input.csv")

df.head()
```

```python
for column in df.columns.values:
    df.loc[df[column] == '-', column] = np.nan

df.head()
```

```python
df.to_csv("output/data_cleaned.csv")
```
